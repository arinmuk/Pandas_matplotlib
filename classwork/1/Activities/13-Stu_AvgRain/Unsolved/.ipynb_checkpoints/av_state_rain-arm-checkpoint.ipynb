```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
file="../resources/avg_rain_state.csv"
rain_df=pd.read_csv(file)
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis=np.arange(len(rain_df["State"]))
#print(x_axis)
x_ticks=[value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis,rain_df["Inches"], color='r', alpha=0.5, align="edge")
plt.xticks(x_ticks,(state for state in rain_df["State"]),rotation="vertical")
```

```python
# Set x and y limits
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(rain_df["Inches"])+10)
```
