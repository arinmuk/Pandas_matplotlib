```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0, 10, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
plt.hlines(0,0,10,alpha=.25)
```

```python
plt.plot(x_axis,sin,marker="o",color="b",label="sine")
plt.plot(x_axis,cos,marker="+",color="r",label="cosine")
```

```python
plt.title("sine and cos curves")
plt.xlabel("Input sampled Real number 0 to 10 ")
plt.ylabel("Value of sin (blue) cos (red) ")
plt.grid()
plt.legend()
plt.legend(loc="lower left")
plt.xlim(0,10)
plt.ylim(-1,1)
plt.show()
```
