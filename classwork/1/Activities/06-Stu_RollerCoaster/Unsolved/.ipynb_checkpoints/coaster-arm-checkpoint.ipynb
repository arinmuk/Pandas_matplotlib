```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.arange(0,130,10)
speed_chain = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
speed_launch = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
ab,=plt.plot(time,speed_chain,color='b', label="dangerdrop")
xy,=plt.plot(time,speed_launch, color='r', label="railgun")
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
```

```python
plt.legend(handles=[ab,xy],loc="best")
plt.grid()
```

```python
plt.show()
```
